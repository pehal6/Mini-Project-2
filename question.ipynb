```python
L = ["Ram", 1, "Shyam", 2, "Aman", 3]

# Separate strings and numbers
strings = sorted([i for i in L if type(i) == str])
numbers = sorted([i for i in L if type(i) == int])

# Combine them
sorted_list = strings + numbers
print("Sorted List:", sorted_list)
```
